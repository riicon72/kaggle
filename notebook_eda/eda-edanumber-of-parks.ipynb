```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<a id="1"></a>
# <p style="background-image: url(https://i.postimg.cc/nzGzm1Xm/publicdomainq-0074997.jpg);font-family:Hiragino Sans;font-size:120%;color:white;text-align:center;border-radius:15px 50px; padding:7px">  Sample data acquisition　</p>

```python
df = pd.read_csv('/kaggle/input/edanumber-of-parks-and-visits/Number of Parks and Visits in Japan.csv')
df.head(5)
```

<a id="1"></a>
# <p style="background-image: url(https://i.postimg.cc/nzGzm1Xm/publicdomainq-0074997.jpg);font-family:Hiragino Sans;font-size:120%;color:white;text-align:center;border-radius:15px 50px; padding:7px"> -visits-　　bar graph　</p>

```python
x = df['year']
y1 = df['national park_visits']
y2 = df['quasi-national park_visits']
y3 = df['prefectural natural park_visits']

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

<a id="1"></a>
# <p style="background-image: url(https://i.postimg.cc/nzGzm1Xm/publicdomainq-0074997.jpg);font-family:Hiragino Sans;font-size:120%;color:white;text-align:center;border-radius:15px 50px; padding:7px"> national park　　park×visits　</p>

```python
df_y = df.set_index('year')
```

```python
fig, ax1 = plt.subplots(1,1,figsize=(10,8))
ax2 = ax1.twinx()
ax1.bar(df_y.index,df["national park"],color="lightblue",label="national park")
ax2.plot(df_y["national park_visits"],linestyle="solid",color="k",marker="^",label="national park_visits")
ax1.set_ylim(0,35)
ax2.set_ylim(1000,500000)
handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
ax1.legend(handler1+handler2,label1+label2,borderaxespad=0)
ax1.grid(True)
fig.show()
```

<a id="1"></a>
# <p style="background-image: url(https://i.postimg.cc/nzGzm1Xm/publicdomainq-0074997.jpg);font-family:Hiragino Sans;font-size:120%;color:white;text-align:center;border-radius:15px 50px; padding:7px"> quasi-national　　park×visits　</p>

```python
fig, ax1 = plt.subplots(1,1,figsize=(10,8))
ax2 = ax1.twinx()
ax1.bar(df_y.index,df["quasi-national park"],color="lightblue",label="quasi-national park")
ax2.plot(df_y["quasi-national park_visits"],linestyle="solid",color="k",marker="^",label="quasi-national park_visits")
ax1.set_ylim(0,65)
ax2.set_ylim(1000,350000)
handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
ax1.legend(handler1+handler2,label1+label2,borderaxespad=0)
ax1.grid(True)
fig.show()
```
