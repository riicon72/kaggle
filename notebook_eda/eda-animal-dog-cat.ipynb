```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_animal = pd.read_csv('/kaggle/input/animalanimal-protection-and-culling/Animal protection and culling.csv')
data_animal
```

# --Details--

```python
a = ['adopt a stray dog','adopt a stray cat','euthanize dog','euthanize cat']
for b in a:
    data_animal[b].unique()
    print('-----'+b+'-------')
    print(data_animal[b].describe())
    print(type(data_animal[b]))
```

# â‘ --ALL--line graph

```python
X_1 = data_animal['year']
Y_1 = data_animal.iloc[:, 1] 
Y_2 = data_animal.iloc[:, 2] 
Y_3 = data_animal.iloc[:, 3] 
Y_4 = data_animal.iloc[:, 4] 

 
plt.plot(X_1, Y_1,'g')
plt.plot(X_1, Y_2,'b')
plt.plot(X_1, Y_3,'r')
plt.plot(X_1, Y_4,'y')

plt.legend()
plt.show()
```
