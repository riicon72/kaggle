```python
ber_df = pd.pivot_table(df, index='gender', columns='Frequency of weighing',  aggfunc=np.sum)
ber_df.plot.bar()
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

```python
ber_df = pd.pivot_table(df, index='age', columns='Frequency of weighing',  aggfunc=np.sum)
ber_df.plot.bar()
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

```python
ber_df = pd.pivot_table(df, index='Frequency of weighing', columns='gender',  aggfunc=np.sum)
ber_df.plot.bar()
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```
