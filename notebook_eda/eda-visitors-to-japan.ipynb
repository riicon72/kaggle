```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/countries-and-regions-of-foreign-visitors-to-japan/2023 Breakdown by country and region_.csv')
df
```

# -Details-

```python
df.columns
```

```python
df.describe()
```

# country-bar graph-

```python
left = df['Country/Area(23 Markets)']
df = df.sort_values([ 'Visitor Arrivals','Country/Area(23 Markets)','Area(23 Markets)','Year'])
height = df['Visitor Arrivals']
plt.barh(left, height, color='b')
plt.show()
```

# Area-bar graph-

```python
df_drop = df.drop(['Country/Area(23 Markets)', 'Year'], axis=1)
grouped = df_drop.groupby('Area(23 Markets)')
group_sums = grouped.sum()
group_sum = group_sums.reset_index()
group_sum
```

```python
left = group_sum['Area(23 Markets)']
group_sum = group_sum.sort_values([ 'Visitor Arrivals','Area(23 Markets)'])
height = group_sum['Visitor Arrivals']
plt.barh(left, height, color='g')
plt.show()
```

# -pie graph-

```python
group_sums
group_sums.plot.pie(y='Visitor Arrivals')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```
